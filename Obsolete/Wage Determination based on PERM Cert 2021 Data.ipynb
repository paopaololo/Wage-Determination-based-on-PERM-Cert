# Reading Dataset


```python
import pandas as pd
import numpy as np
#pd.options.mode.chained_assignment = None # default='warn' Get rid of SettingWithCopyWarning. Not recommended during data cleaning. Enabled only for the prediction. 
```

```python
df = pd.read_csv ("PERM_Disclosure_Data_FY2021.csv", low_memory=False)
df
```

```python
#Extracting the vital columns to a new Dataframe:
extract_df = df.loc[:, ["CASE_STATUS", "PW_SOC_CODE", "PW_SOC_TITLE", "PW_SKILL_LEVEL", "PW_WAGE", 
                    "PW_UNIT_OF_PAY", "WAGE_OFFER_FROM", "WAGE_OFFER_TO", "WAGE_OFFER_UNIT_OF_PAY",
                    "WORKSITE_STATE", "WORKSITE_POSTAL_CODE", "FOREIGN_WORKER_EDUCATION", "WORKSITE_CITY", "REQUIRED_EXPERIENCE", "FOREIGN_WORKER_REQ_EXPERIENCE"]]
```

```python
extract_df
```

# Data Cleaning
1. Filter for "Certified-Expired" and "Certified"
    We do not need distinguish between "Certified-Expired" and "Certified". They are both approved by the labor department.
2. Filter for Full-time Yearly Salary jobs
3. Cleaning and Manipulating the wages
    Change the prevailing wage back to float. Taking the midpoint of the wage offer to target_wage.
4. Cleaning postal code
    Remove numbers after "-". Remove non numerical values. Leave them as string instead of numerical.
5. Cleaning PW SOC
    Remove all NaNs and convert everything into string. Each code represent a job class. It cannot be numerical
6. Clean up missing values for remaining fields

```python
# Step 1
# Filter for "Certified-Expired" and "Certified"
list = ['Certified-Expired', 'Certified'] 
extract_df = extract_df[extract_df['CASE_STATUS'].isin(list)]
```

```python
# Step 2
# Filter for Full-time Yearly Salary jobs
extract_df = extract_df.loc[(extract_df['PW_UNIT_OF_PAY'] == "Year")]
extract_df = extract_df.loc[(extract_df['WAGE_OFFER_UNIT_OF_PAY'] == "Year")]
```

```python
# Step 3
# Taking the midpoint of the wage offer


target_wage = []
for index, row in extract_df.iterrows():
    row['PW_WAGE'] = row['PW_WAGE'].replace('$','').replace(',','')
    row['PW_WAGE'] = float(row['PW_WAGE'])
    row['WAGE_OFFER_FROM'] = row['WAGE_OFFER_FROM'].replace('$','').replace(',','')
    row['WAGE_OFFER_FROM'] = float(row['WAGE_OFFER_FROM'])
    if isinstance(row['WAGE_OFFER_TO'], str) :
            row['WAGE_OFFER_TO'] = row['WAGE_OFFER_TO'].replace('$','').replace(',','')
            row['WAGE_OFFER_TO'] = float(row['WAGE_OFFER_TO'])
            target_wage.append(row['WAGE_OFFER_FROM'] 
                               + (row['WAGE_OFFER_TO'] - row['WAGE_OFFER_FROM'])/2)
    else:
            target_wage.append(row['WAGE_OFFER_FROM'])

extract_df['Target_Wage']= target_wage
extract_df['PW_WAGE'] = extract_df['PW_WAGE'].apply(float)
extract_df['Target_Wage'] = extract_df['Target_Wage'].apply(float)

extract_df
    
```

```python
# Step 4
# Cleaning postal code
import re #import Regex

for index, row in extract_df.iterrows():
    if  re.search('[a-zA-Z]', row['WORKSITE_POSTAL_CODE']): # Check if the zip contains any alphabet
        extract_df.drop(row.name, inplace=True)    
    if '-' in row['WORKSITE_POSTAL_CODE']: # Strip andy ZIP+4 to Five-Digit ZIP
        temp = row['WORKSITE_POSTAL_CODE'].split('-')[:1]
        extract_df.at[row.name, 'WORKSITE_POSTAL_CODE'] = temp #Avoid SettingWithCopyWarning 
extract_df['WORKSITE_POSTAL_CODE'] = extract_df['WORKSITE_POSTAL_CODE'].apply(str) #force all postal codes to be strings
```

```python
# Step 5
# Cleaning PW_SOC_CODE
extract_df['PW_SOC_CODE'] = extract_df['PW_SOC_CODE'].apply(str) #force all SOC codes to be strings
for index, row in extract_df.iterrows(): #check if there are irregular SOC codes
    if not '-' in row['PW_SOC_CODE']:
        extract_df.at[row.name, 'PW_SOC_CODE'] = ''
        # print(row['PW_SOC_CODE'])
extract_df.dropna(subset=['PW_SOC_CODE']) #drop all NaN for this field
```

```python
# Step 6

extract_df['PW_SKILL_LEVEL'] = extract_df['PW_SKILL_LEVEL'].replace(['N/A'], '')

# Other data are outliers in this case
extract_df['FOREIGN_WORKER_EDUCATION'] = extract_df['FOREIGN_WORKER_EDUCATION'].replace(['Other'], '') 

#drop every vital attributes with null values
extract_df = extract_df.dropna(subset=[
    'PW_SKILL_LEVEL', 'Target_Wage', 'WORKSITE_POSTAL_CODE',
    'PW_WAGE', 'FOREIGN_WORKER_EDUCATION', 'PW_SOC_CODE'
    ])

extract_df.isnull().sum()
```

```python
extract_df
```

# Data Transformation
Encode all the categorical columns 

```python
from sklearn.preprocessing import OrdinalEncoder

#FOREIGN_WORKER_EDUCATION: need to code it manually
temp_label = {'None' : 0,
             'High School' : 1,
             "Associate's" : 2,
             "Bachelor's": 3,
             "Master's" : 4,
             "Doctorate" :5}
extract_df['FOREIGN_WORKER_EDUCATION_ORD'] = extract_df.FOREIGN_WORKER_EDUCATION.map(temp_label).copy()

#PW_SOC_CODE, #PW_SKILL_LEVEL, #WORKSITE_POSTAL_CODE: this is treated as string instead of number
enc = OrdinalEncoder()
extract_df[["PW_SOC_CODE_ORD",
            "PW_SKILL_LEVEL_ORD", 
            "WORKSITE_POSTAL_CODE_ORD",
            "REQUIRED_EXPERIENCE_CD",
            "FOREIGN_WORKER_REQ_EXPERIENCE_CD"]] = enc.fit_transform(extract_df[["PW_SOC_CODE",
                                                                              "PW_SKILL_LEVEL", 
                                                                              "WORKSITE_POSTAL_CODE",
                                                                              "REQUIRED_EXPERIENCE",
                                                                              "FOREIGN_WORKER_REQ_EXPERIENCE"
                                                                             ]]).copy()



extract_df = extract_df.dropna(subset=[
    'PW_SKILL_LEVEL_ORD',
    'FOREIGN_WORKER_EDUCATION_ORD',
    'REQUIRED_EXPERIENCE_CD',
    'FOREIGN_WORKER_REQ_EXPERIENCE_CD'
    ])
extract_df.isnull().sum()
```

```python
extract_df
```

# Splitting the Data

```python
from sklearn.model_selection import train_test_split
cols_to_use = ['PW_SOC_CODE_ORD', 'PW_SKILL_LEVEL_ORD', 'WORKSITE_POSTAL_CODE_ORD', 'FOREIGN_WORKER_EDUCATION_ORD', 'PW_WAGE']
X = extract_df[cols_to_use]
y = extract_df.Target_Wage
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                                random_state=5)
```

# Building the Model

```python
import xgboost as xgb
import matplotlib.pyplot as plt
import graphviz

my_model = xgb.XGBRegressor(n_estimators=450, learning_rate=0.5, n_jobs=16)
my_model.fit(X_train, y_train)
print(my_model)

#plotting the decision tree
xgb.plot_tree(my_model, num_trees=2)
fig = plt.gcf()
fig.set_size_inches(250, 200)
fig.savefig('tree.png')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
#validate the data and calculate MAE
predictions = my_model.predict(X_valid)
mae = mean_absolute_error(predictions, y_valid)
rsme = mean_squared_error(predictions, y_valid, squared=False)
print('Mean:', y.mean(), 'Median:', y.median())
print("Mean Absolute Error:" , mae)
print("Root Mean Square Error :" , rsme)
```

Since RMSE > MAE, the result means we have larger variance of the frequency distribution of errors magnitudes.

## Ploting the map using folium

```python
#load the geo data and the salary level
folium.Choropleth(geo_data="nyc-zip-code-tabulation-areas-polygons.geojson",
               data=NYnearby_df, 
               columns=['WORKSITE_POSTAL_CODE', 'Target_Wage'], 
               key_on='feature.properties.postalCode', 
               fill_color='YlOrRd',fill_opacity=0.7,line_opacity=0.2,
               legend_name='Salary level').add_to(NYmap)
NYmap
```
